```python
import time
import torch
import torch.nn as nn
from NN_scratch import Linear 
def test_linear_layer():
    print("=== TEST LỚP LINEAR (FULLY CONNECTED) ===")
    
    batch_size = 32       # Xử lý cùng lúc 32 bức ảnh/mẫu dữ liệu
    in_features = 512     # Kích thước đầu vào (Ví dụ: mảng đã được duỗi thẳng)
    out_features = 128    # Kích thước đầu ra (Giảm chiều dữ liệu)
    
    # 1. Khởi tạo dữ liệu giả
    x_numpy = np.random.randn(batch_size, in_features)
    x_torch = torch.tensor(x_numpy, dtype=torch.float32)
    
    # 2. Khởi tạo 2 mô hình
    model_custom = Linear(in_features, out_features)
    model_torch = nn.Linear(in_features, out_features)
    
    # 3. Đồng bộ trọng số (Ép PyTorch dùng trọng số của NumPy)
    with torch.no_grad():
        model_torch.weight.copy_(torch.tensor(model_custom.weights, dtype=torch.float32))
        model_torch.bias.copy_(torch.tensor(model_custom.bias, dtype=torch.float32))
        
    # 4. Chạy thử NumPy
    start_time = time.time()
    out_custom = model_custom.forward(x_numpy)
    time_custom = time.time() - start_time
    
    # 5. Chạy thử PyTorch
    start_time = time.time()
    with torch.no_grad():
        out_torch = model_torch(x_torch)
    time_torch = time.time() - start_time
    
    # 6. So sánh
    print(f"Kích thước Input : {x_numpy.shape} -> (Batch_size, In_features)")
    print(f"Kích thước Output: {out_custom.shape} -> (Batch_size, Out_features)")
    
    # So sánh độ chính xác
    # Chuyển tensor PyTorch về lại numpy để so sánh
    is_correct = np.allclose(out_custom, out_torch.numpy(), atol=1e-5)
    
    print("\n--- KẾT QUẢ ---")
    if is_correct:
        print("✅ ĐỘ CHÍNH XÁC: Hoàn hảo! Toán học của bạn hoàn toàn khớp với PyTorch.")
    else:
        print("❌ ĐỘ CHÍNH XÁC: Có sự sai lệch.")
        
    print(f"⏳ Thời gian NumPy  : {time_custom:.6f} giây")
    print(f"⏳ Thời gian PyTorch: {time_torch:.6f} giây")

# Chạy test
test_linear_layer()
```
