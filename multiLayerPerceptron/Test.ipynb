```python
import torch
def train(model, x_data, y_target, learning_rate=0.01, epochs=100):
    for epoch in range(epochs):
        
        prediction = model.forward(x_data)
        
        
        
        loss = torch.mean((prediction - y_target)**2)
        

        
        grad_loss = 2 * (prediction - y_target) / x_data.shape[0]
        
        
        model.backward(grad_loss, learning_rate)

        if (epoch + 1) % 10 == 0:
            print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.6f}")
```

```python
from MLP import MLP
import numpy as np 

# Get input and output dimensions from training data
input_dim = train_data.shape[1]  # number of features
output_dim = 1  # for regression/binary classification, adjust if needed

# Convert numpy arrays to torch tensors
X_train = torch.tensor(train_data[:, :-1].astype(np.float32))  # features (all but last column)
y_train = torch.tensor(train_data[:, -1:].astype(np.float32))  # target (last column)

# Initialize the model
model = MLP(dim_in=input_dim-1, dim_out=output_dim, lr = 0.0001)

# Train the model
train(model, X_train, y_train, learning_rate=0.01, epochs=100)
```
